```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a waveform
fs = 100
f1 = 2.0
f2 = 3.0
t = np.arange(0, 10, 1./fs)
x1 = 1.34 * np.sin(2*np.pi*f1*t)
x2 = 2.55 * np.cos(2*np.pi*f2*t)
x  = x1 + x2
# Plot the waveform
plt.plot(t, x);plt.xlabel('Time (s)');plt.ylabel('Amplitude');
```

```python
# Design a peak filter
Q = 30.0 # Quality factor
b, a = signal.iirpeak(f2, Q, fs)

# Filter the waveform with gust method
filtered_x = signal.filtfilt(b, a, x, method='gust')

# Plot and compare the waveforms
plt.figure(facecolor='white')
plt.plot(t, x, 'gray', linestyle='dashed', lw=1.0, label='Mixed Waveform')
plt.plot(t, x2, 'red', linestyle='dashdot', lw=1.0, label='Raw Waveform')
plt.plot(t, filtered_x, 'blue', linestyle='dashed', lw=1.0, label='Narrow Bandpassed Waveform')
plt.xlabel('Time (s)');plt.ylabel('Amplitude')
plt.xlim(0, 10);plt.ylim(-5, 5)
plt.legend(loc='upper right');plt.tight_layout()
```
